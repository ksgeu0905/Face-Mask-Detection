```python
import numpy as np
import cv2
```

```python
with_mask = np.load('with_mask.npy')
without_mask = np.load('without_mask.npy')
```

```python
with_mask.shape
```

```python
without_mask.shape
```

```python
#Reshaping our data so that analysis of the feature is easier
with_mask=with_mask.reshape(200,50*50*3)
without_mask=without_mask.reshape(200,50*50*3)
```

```python
with_mask.shape
```

```python
X=np.r_[with_mask,without_mask] #concatinating both the datasets
```

```python
X.shape
```

```python
lables = np.zeros(X.shape[0])
```

```python
lables[200:]=1.0
```

```python
names = {0: 'Mask' , 1: 'No Mask'}
```

```python
# svm= Support Vector Machine
# SVC= Support Vector Classification
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split (X,lables,test_size=0.25)
```

```python
# PCA= Principal component Analysis
# PCA, is a dimensionality-reduction method that is often used to reduce the dimensionality of large data sets.
from sklearn.decomposition import PCA
```

```python
pca=PCA(n_components=3)
x_train=pca.fit_transform(x_train)
```

```python
x_train.shape
```

```python
svm=SVC()
# implementing the algorithm on the training data
svm.fit(x_train,y_train)
```

```python
# transforming the dimensions of the testing data and predicting the result after training the algorithm..
x_test=pca.transform(x_test)
y_pred=svm.predict(x_test)
```

```python
haar_data=cv2.CascadeClassifier("C:\\Users\\HP\\Desktop\\face_mask_easy\\haarcascade_frontalface_default.xml")
capture=cv2.VideoCapture(0)
data=[]
font=cv2.FONT_HERSHEY_COMPLEX
while True:
    flag,img= capture.read()
    if flag:
        faces=haar_data.detectMultiScale(img)
        for x,y,w,h in faces:
            face=img[y:y+h,x:x+w,:] #slicing face out of the frame
            face=cv2.resize(face,(50,50)) # resizing the face to 50 * 50
            face=face.reshape(1,-1) 
            face=pca.transform(face) 
            pred=svm.predict(face) # predicting the result on each frame
            n=names[int(pred)] #converting the result into an integer
            if int(pred)==0:
                cv2.rectangle(img,(x,y),(x+w,y+h),(0,255,0),4) #this will execute if the person is wearing mask
            else:
                cv2.rectangle(img,(x,y),(x+w,y+h),(0,0,255),4) # this will execute if the person not wearing mask
            cv2.putText(img,n,(x,y),font,1,(244,250,250),2) # putting the text on the output screen
            print(n)
        cv2.imshow('result',img)
        #27 for escape ascii
        if cv2.waitKey(2) == 27:
            break
capture.release()
cv2.destroyAllWindows()
```
